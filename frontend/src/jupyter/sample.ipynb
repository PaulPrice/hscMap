```python
import numpy

# カタログをオーバープロット
def radec(c, *select):
    select = reduce(lambda a,b: numpy.logical_and(a, b), select)
    return numpy.array([c.ra[select], c.dec[select]]).T

c = cat2
v = numpy.array([c.imag_kron, c.rmag_kron, c.gmag_kron]).max(axis=0)

w.clearMarkers()
colors = [[1, 0, 0, 0.5],
          [1, 1, 0, 0.5],
          [0, 1, 0, 0.5],
          [0, 1, 1, 0.5],
          [0, 0, 1, 0.5],
          [1, 0, 1, 0.5]]

w.clearMarkers()
bright = 24
for i in range(5):
    w.addMarker(radec(c, bright + i <= v, v < bright + 1 + i).tolist()).color = colors[i]
w.addMarker(radec(c, 29 <= c.imag_kron).tolist()).color = colors[5]
```

```python
# オーバープロットを消す
w.clearMarkers()
```
