````python
import numpy
coords_s = '''
36.54        -4.969          0.054
36.378       -4.239          0.142
36.917       -4.858           0.29
37.012         -4.851           0.26
36.017         -4.225           1.05
34.908         -4.007            0.2
34.608         -4.997           1.13
35.834         -5.454           0.21
34.632         -5.017           0.87
36.635         -5.001           0.49
35.065         -4.917           0.51
36.387         -5.539           0.57
35.64         -4.967           0.83
'''
a = numpy.array(map(float, coords_s.split()))
a = a.reshape((len(a)/3,3))
````

````python
w.udeep = w.deep = False
w.wide=True
````

````python
w.focusTable(['ra','dec','z'], a)
````

````python
import subprocess
from astropy.io import fits as afits

w.clearContours()

for i, (ra, dec, z) in enumerate(a):
    subprocess.check_call(
        'python crop.py --out {out}.fits --alpha {alpha} --delta {delta} --fov 30 Map.XMM.fits'.format(
            alpha=ra,
            delta=dec,
            out=i,
        ),
        shell=True)
    with afits.open('./{}.fits'.format(i)) as hdul:
        w.addContour(hdul[1], levels=numpy.arange(0,5,0.5))
````
